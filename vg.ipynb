# Análise de dados para vendas de Video Games e jogos 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#jogos até 2016
#1 - SERIE COM MAIOR VENDA GLOBAL
#2 - Dez maiores vendas
#3 - O mesmo em forma de lista
#4 - Contagem das empresas que mais vendem
#5 - Empresas mais lucrativas
#6 - Gêneros mais populares
#7 - Plataformas mais populares
#8 - Plataformas mais lucrativas 
df = pd.read_csv('vgsales.csv')
```

```python
df[df['Global_Sales'] == df['Global_Sales'].max()] #serie do elemento com maior venda global
```

```python
gg = df.set_index('Name')
gg['Global_Sales'].nlargest(10)
```

10 maiores vendas globais

```python
c = df.groupby('Name')['Global_Sales'].sum() #relacionando vendas globais e jogos
list(c.nlargest(5).index) #5 jogos mais vendidos no mundo em 
```

```python
d = df.groupby('Publisher')['Global_Sales'].count() #
d.nlargest(10)
```

Contagem das empresas que mais vendem

```python
e = df.groupby('Publisher')['Global_Sales'].sum()
e.nlargest(5) #empresas mais lucrativas
```

Empresas produtoras de jogos com maior lucratividade 

```python
f = df.groupby('Genre')['Global_Sales'].count() #generos mais populares
f.nlargest(5)
```

Gêneros de jogos mais vendidos

```python
g = df.groupby('Platform')['Global_Sales'].count() #contagem das plataformas mais populares
g.nlargest(5)
```

Plataformas mais vendidas 

```python
h = df.groupby('Platform')['Global_Sales'].sum() #contagem das plataformas mais lucrativas
h.nlargest(5)
```

Distribuição Regional

```python
#distribuição percentual por regiõa
df['NA %'] = 100*df['NA_Sales']/df['Global_Sales']
df['EU %'] = 100*df['EU_Sales']/df['Global_Sales']
df['JP %'] = 100*df['JP_Sales']/df['Global_Sales']
df['Others %'] = df['Other_Sales']/df['Global_Sales']
```

```python
kkk = df[df['Platform'] == 'N64']
kkk = kkk.sort_values(by=['Global_Sales'],ascending=False)
x = np.arange(1,len(kkk)+1)
kkk = kkk.drop(['Rank','Platform','JP_Sales','NA_Sales','EU_Sales','Other_Sales','NA %','EU %','JP %','Others %'],axis=1)
jjj = kkk.sort_values(by=['Global_Sales'],ascending=False)
jjj['Sales Position'] = x
jjj = jjj.set_index('Name')
jjj['% Total Sales N64'] = 100*jjj['Global_Sales']/sum(jjj['Global_Sales'])
jjj.head(20)
```

20 maiores vendas de N64

```python
vc = jjj.groupby('Publisher')['% Total Sales N64'].sum()
cv = jjj.groupby('Publisher')['% Total Sales N64'].count()
ef = vc/cv
vc.nlargest(10) #vendas totais das produtoras
```
